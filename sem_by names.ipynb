```python
import pandas as pd
import re
import gensim
import logging
import nltk.data
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from gensim.models import word2vec
from pymystem3 import Mystem
import warnings
warnings.filterwarnings('ignore')
```

```python
table = pd.read_excel('fooddate.xlsx')
```

```python
table
```

```python
m = Mystem()
```

```python
name_type = []
for index, row in table.iterrows():
    name_ = row['Name']
    name = re.sub(r'[^\w\s]','',name_)
    name = re.sub(r'[0-9]', '', name)
    type_obj = row['TypeObject']
    nam_lem = m.lemmatize(name)
    print(nam_lem[:-1])
    line = [nam_lem[:-1], type_obj]
    name_type.append(line)
```

```python
for name_t in name_type:
    line_ = ''.join(name_t[0])
    line = str(line_) + ' '+ name_t[1]
    f = open('text_name_obj.txt', 'a')
    f.write(str(line) + '\n')
    f.close()
```

```python
f = 'text_name_obj.txt'
data = gensim.models.word2vec.LineSentence(f)
```
