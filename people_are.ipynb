```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
table = pd.read_excel('fooddate.xlsx')
```

```python
table
```

```python
pivot_adm_area = table.pivot_table(index = 'AdmArea', values = 'Name', aggfunc='count')
```

```python
s = [154.84, 153.03, 37.20, 361.36, 113.73, 101.88, 93.28, 1084.34, 66.18, 117.56, 111.36, 131.77]
pivot_adm_area['Площадь округа (км^2)'] = s
people = [1524265, 1395986, 250173, 275508, 1186128, 1432571, 1011387, 130315, 779352, 1431746, 1446432, 1791187]
pivot_adm_area['Население округа (чел)'] = people
```

```python
pivot_adm_area
```

```python
count_s = []
for index, row in pivot_adm_area.iterrows():
    x = round(row['Name']/row['Площадь округа (км^2)'], 1)
    n_s = [index, x]
    count_s.append(n_s)
#Заведений на квадратный километр
df1 = pd.DataFrame(count_s, columns = ['Округ', 'Заведений на квадратный километр'])
df1
```

```python
count_p = []
for index, row in pivot_adm_area.iterrows():
    x = round(row['Население округа (чел)']/row['Name'], 1)
    n_s = [index, x]
    count_p.append(n_s)    
#Количество человек на одно заведение
df2 = pd.DataFrame(count_p, columns = ['Округ', 'Количество человек на одно заведение'])
df2
```
