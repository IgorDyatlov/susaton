```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
table = pd.read_excel('fooddate.xlsx')
```

```python
table
```

```python
pivot_district = table.pivot_table(index = 'District', values = 'Name', aggfunc='count')
```

```python
pivot_district.sort_values('Name')
```

```python
pivot_district.sort_values('Name').describe()
```

* Наименьшее количестве заведений в раоне: 2
